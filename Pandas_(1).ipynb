```python
import pandas as pd
import io

# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}

df = pd.DataFrame(data)

# Capture the output of df.info()
buffer = io.StringIO()
df.info(buf=buffer)
info_str = buffer.getvalue()

# Convert the info string into a DataFrame
info_df = pd.DataFrame({'Info': info_str.split('\n')})

# Write both the df and info_df to separate sheets in an Excel file
excel_file_path = 'df_and_info.xlsx'
with pd.ExcelWriter(excel_file_path, engine='openpyxl') as writer:
    # Write the DataFrame to the first sheet
    df.to_excel(writer, sheet_name='DataFrame', index=False)
    # Write the info DataFrame to the second sheet
    info_df.to_excel(writer, sheet_name='DataFrame Info', index=False)

print(f"DataFrame and its info have been written to {excel_file_path}")
```

```python
# The groupby() function is a useful tool for grouping and aggregating data.
# Here are a few examples of how to use the groupby() function in Pandas:

# Grouping data by a single column:

import pandas as pd

# Create a sample dataframe
data = {'name': ['Alice', 'Bob', 'Charlie', 'Bob', 'Charlie'],
        'age': [25, 30, 35, 40, 45],
        'city': ['New York', 'Los Angeles', 'Chicago', 'New York', 'Chicago']}
df = pd.DataFrame(data)
print(df)
# Group the data by the 'name' column
grouped = df.groupby('name')

#Print the groups
print(grouped.groups)
```

```python
# Aggregating data after grouping:
import pandas as pd

# Create a sample dataframe
data = {'name': ['Alice', 'Bob', 'Charlie', 'Bob', 'Charlie'],
        'age': [25, 30, 35, 40, 45],
        'city': ['New York', 'Los Angeles', 'Chicago', 'New York', 'Chicago'],
        'income': [50000, 60000, 55000, 60000, 70000]}
df = pd.DataFrame(data)
print(df)
# Group the data by the 'name' column and compute the mean of the 'income' column
grouped = df.groupby('name')['income'].mean()

# Print the result
print(grouped)
```

```python
# Applying a custom function to groups:
import pandas as pd

# Create a sample dataframe
data = {'name': ['Alice', 'Bob', 'Charlie', 'Bob', 'Charlie'],
        'age': [25, 30, 35, 40, 45],
        'city': ['New York', 'Los Angeles', 'Chicago', 'New York', 'Chicago'],
        'income': [50000, 60000, 55000, 60000, 70000]}
df = pd.DataFrame(data)
print(df)


# Define a custom function to compute the range of ages
def age_range(group):
    return group.max() - group.min()

# Group the data by the 'name' column and apply the custom function to the 'age' column
grouped = df.groupby('name')['age'].apply(age_range)

# Print the result
print(grouped)
```

```python
# Using transform function
import pandas as pd

# Create a sample dataframe
data = {'name': ['Alice', 'Bob', 'Charlie', 'Bob', 'Charlie'],
        'age': [25, 30, 35, 40, 45],
        'city': ['New York', 'Los Angeles', 'Chicago', 'New York', 'Chicago'],
        'income': [50000, 60000, 55000, 60000, 70000]}
df = pd.DataFrame(data)

# Group the data by the 'city' column and use transform to add a new column with the mean income of each city
df['mean_income'] = df.groupby('city')['income'].transform('mean')

# Print the dataframe
print(df)
```

```python
# Using first() and last() function
import pandas as pd

# Create a sample dataframe
data = {'name': ['Alice', 'Bob', 'Charlie', 'Bob', 'Charlie'],
        'age': [25, 30, 35, 40, 45],
        'city': ['New York', 'Los Angeles', 'Chicago', 'New York', 'Chicago'],
        'income': [50000, 60000, 55000, 60000, 70000]}
df = pd.DataFrame(data)

# Group the data by the 'name' column and get the first and last item of each group
first_item = df.groupby('name').first()
last_item = df.groupby('name').last()

# Print the result
print(first_item)
print(last_item)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(0)

# Generate irregular and stationary time series data
n = 100  # Number of data points
irregular_component = np.random.normal(loc=0, scale=0.2, size=n)  # Irregular component
stationary_component = np.sin(np.linspace(0, 4*np.pi, n))  # Stationary component
timeseries = irregular_component + stationary_component  # Combined time series data

# Create a DataFrame with the time series data
data = pd.DataFrame({'Time': pd.date_range(start='2000-01-01', periods=n), 'Value': timeseries})

# Plot the time series data
plt.figure(figsize=(10, 6))
plt.plot(data['Time'], data['Value'])
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Irregular and Stationary Time Series Data')
plt.show()
```
